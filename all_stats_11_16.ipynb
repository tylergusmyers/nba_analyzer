```python
def count_stats(james_df):
    a = james_df['pts']
    b = james_df['fgm'] * 0.4
    c = james_df['fga'] * 0.7
    d = ((james_df['fta']) - (james_df['ftm'])) * 0.4
    e = james_df['oreb'] * 0.7 
    f = james_df['dreb'] * 0.3
    g = james_df['stl']
    h = james_df['ast'] * 0.7
    j = james_df['blk'] * 0.7
    k = james_df['pf'] * 0.4
    l = james_df['turnover']
    james_score = a + b - c - d + e + f + g + h + j - k - l
    id_name = mapping_id[player_id]
    return pd.DataFrame(james_score, column_name = id_name)
```
