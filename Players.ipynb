```python
import requests
import pandas as pd
import json
import pprint
```

```python
# Read the CSV file containing highest paid NBA players
salary_df = pd.read_csv('Resources/nba_salaries.csv')
```

```python
# Isolate the names of NBA players into a list
player_names_list = salary_df["Player"].values.tolist()
```

```python
# Run the list of names through our API call and parse out player ID's to use later

## For some reason, Kristaps and Devin Booker get dropped from the list...
## It could be a limit on the API call?

url = "https://www.balldontlie.io/api/v1/players/?search="
player_id_list = []

for name in player_names_list:
    player_url = url + f"{name}"
    player_get = requests.get(player_url)
    player_get.json()
    player_dict = json.loads(player_get.text)
    player_id_list.append(player_dict['data'][0]["id"])
```

```python
# Hint --  Give some time some the JSON to run
# Add the information to a DataFrame
player_id_df = pd.DataFrame(player_id_list)
salary_id_join = player_id_df.join(salary_df)
salary_player_id_df = salary_id_join.rename(columns={0: 'player_id'})
salary_player_id_df= salary_player_id_df[['player_id','Player','Tm','2021-22']]
salary_player_id_df
```
